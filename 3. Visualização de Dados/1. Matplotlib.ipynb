## VAMOS CRIAR NOSSO PRIMEIRO GRÁFICO DE BARRAS

* Iremos utilizar o método plot junto ao DataFrame.

* Link do método: [pandas.DataFrame.plot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html)

```
Parameters
data: Series or DataFrame
>>> The object for which the method is called.

x: label or position, default None
>>> Only used if data is a DataFrame.

y: label, position or list of label, positions, default None
>>> Allows plotting of one column versus another. Only used if data is a DataFrame.

kind: str
>>> The kind of plot to produce:

    > 'line': line plot (default)

    > 'bar': vertical bar plot

    > 'barh': horizontal bar plot

    > 'hist': histogram

    > 'box': boxplot

    > 'kde': Kernel Density Estimation plot

    > 'density': same as ‘kde’

    > 'area': area plot

    > 'pie': pie plot

    > 'scatter': scatter plot (DataFrame only)

    > 'hexbin': hexbin plot (DataFrame only)

figsize: a tuple (width, height) in inches
>>> Size of a figure object.

```

````python
df.plot(
    kind = 'bar'
    , x = 'categoria'
    , y = 'vl_produto'
)
````

````python
df.plot(
    kind = 'bar'
    , x = 'categoria'
    , y = 'vl_produto'
    , figsize = (18,5)
)

````

````python
df.plot(
    kind = 'bar'
    , x = 'categoria'
    , y = 'vl_produto'
    , figsize = (18,5)
    , color = 'green'
)

````

````python
ax = df.plot(
    kind = 'bar'
    , x = 'categoria'
    , y = 'vl_produto'
    , figsize = (18,5)
    , color = 'green'
    , rot = 0
)

ax.set_title('Valor de Vendas x Categoria')
ax.set_xlabel('Categorias')
ax.set_ylabel('Valor de Venda')
ax

plt.savefig('Teste.png')
````

````python
import pandas_datareader.data as web
````

````python
df_acoes = web.DataReader(
    'MSFT34.SA'
    , 'yahoo'
    , start='2021-08-23'
    , end='2021-08-27'
    )

df_acoes.head()
````

````python
df_acoes = df_acoes.reset_index()
df_acoes.columns = ['dt_acao', 'alto', 'baixo', 'abertura', 'fechamento', 'volume', 'fechamento_ajustado']
````

````python
df_acoes.head()
````

````python
ax = df_acoes.plot(
    kind = 'line'
    , x = 'dt_acao'
    , y = 'abertura'
    , figsize = (18, 6)
    , marker='o'
)
````

````python
ax = df_acoes.plot(
    kind = 'line'
    , x = 'dt_acao'
    , y = ['abertura', 'fechamento', 'alto']
    , figsize = (18, 6)
    , marker='o'
    , color = ['red', 'green', 'blue']
    , linestyle = '--'
)

ax.set_title('AÇÕES DA MICROSOFT - Abertura x Fechamento x Pico')
ax.set_xlabel('Data')
ax.set_ylabel('Valores')
ax
````

````python
# AAPL

df_apple = web.DataReader(
    'AAPL'
    , 'yahoo'
    , start='2021-07-01'
    , end='2021-07-30'
    )

df_apple = df_apple.reset_index()
df_apple.columns = ['dt_acao', 'alto', 'baixo', 'abertura', 'fechamento', 'volume', 'fechamento_ajustado']
df_apple.head()
````

````python
ax = df_apple.plot(
    kind = 'line'
    , x = 'dt_acao'
    , y = ['abertura', 'fechamento']
    , figsize = (18, 6)
    , marker='o'
    , color = ['orange', 'green']
    , linestyle = '--'
)

ax.set_title('AÇÕES DA APPLE - Abertura x Fechamento x Pico')
ax.set_xlabel('Data')
ax.set_ylabel('Valores')
ax
````

````python
ax = df_apple.plot(
    kind = 'line'
    , x = 'dt_acao'
    , y = 'baixo'
    , figsize = (18, 6)
    , marker='o'
    , color = 'red'
    , linestyle = '--'
)

ax.set_title('AÇÕES DA APPLE - Baixo')
ax.set_xlabel('Data')
ax.set_ylabel('Valores')
ax
````
