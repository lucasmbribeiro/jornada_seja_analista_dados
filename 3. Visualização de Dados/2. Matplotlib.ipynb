__Importa as bibliotecas necessárias para o Notebook.__

````python
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
````

- __Pega os dados das ações da Apple entre os dias 01/08/2021 à 30/08/2020__

- __Quebra o indíce do DataFrame.__

- __Renomeia as colunas do DataFrame.__

````python
df_acoes = web.DataReader(
    'AAPL'
    , 'yahoo'
    , start='2021-08-01'
    , end='2021-08-30'
    )

df_acoes = df_acoes.reset_index()
df_acoes.columns = ['dt_acao', 'alto', 'baixo', 'abertura', 'fechamento', 'volume', 'fechamento_ajustado']
df_acoes.head()
````

- __Plota um gráfico de linha para exibir os dados do DataFrame.__

````python
ax = df_acoes.plot(
    kind = 'line'
    , x = 'dt_acao'
    , y = ['abertura', 'fechamento']
    , figsize = (18, 6)
    , marker='o'
    , color = ['blue', 'green']
    , linestyle = '--'
)

ax.set_title('AÇÕES DA APPLE - Abertura x Fechamento x Pico')
ax.set_xlabel('Data')
ax.set_ylabel('Valores')
ax
````

````python
# 1. DataFrame
# 2. Colunas do eixo X
# 3. Colunas do eixo Y
# 4. Marcador
# 5. Cores
# 6. Estilo da linha

def monta_grafico_linha(df, eixo_x, eixo_y, cores = None):
    ax = df.plot(
        kind = 'line'
        , x = eixo_x
        , y = eixo_y
        , figsize = (18, 6)
        , marker='o'
        , color = cores
        , linestyle = '--'
    )

    return ax

````

````python
monta_grafico_linha(df_acoes, 'dt_acao', ['alto', 'baixo'], ['purple', 'black'])
````
