```python
import pandas as pd
```

### Base olist_orders_dataset
*** 

Informações globais do pedido.

|Coluna | Descrição |
| :--- | :------- |
|order_id|id do pedido |
|customer_id|Identificação do Cliente|
|order_status|status do pedido|
|order_purchase_timestamp|registro de data e hora de compra do pedido|
|order_approved_at|pedido aprovado em|
|order_delivered_carrier_date|data da transportadora do pedido entregue|
|order_delivered_customer_date|pedido entregue data do cliente|
|order_estimated_delivery_date|data de entrega estimada do pedido|

```python
df_orders = pd.read_csv("./base/olist_orders_dataset.csv")
df_orders.head()
```

```python
df_orders.rename(columns={
    "order_id": "id_pedido "
    ,"customer_id": "id_cliente"
    ,"order_status": "status_pedido"
    ,"order_purchase_timestamp": "dt_compra_pedido"
    ,"order_approved_at": "dt_pedido_aprovado"
    ,"order_delivered_carrier_date": "dt_transportadora_pedido_entregue"
    ,"order_delivered_customer_date": "dt_entregue_cliente"
    ,"order_estimated_delivery_date": "dt_entrega_estimada_pedido"
}, inplace=True)
```

### Base olist_order_items_dataset
*** 

* Informações dos itens do pedido. 
* É como se fosse os itens do carrinho da loja. 
* Um pedido pode ter mais de um item.

|Coluna | Descrição |
| :--- | :------- |
|order_id|id do pedido |
|order_item_id|ID do item do pedido|
|product_id|ID do produto|
|seller_id|id do vendedor|
|shipping_limit_date|data limite de envio|
|price|preço|
|freight_value|valor do frete|

```python
df_orders_items = pd.read_csv("./base/olist_order_items_dataset.csv")
df_orders_items.head()
```

```python
df_orders_items.rename(columns={
    "order_id": "id_pedido"
    ,"order_item_id": "id_item_pedido"
    ,"product_id": "id_produto"
    ,"seller_id": "id_vendedor"
    ,"shipping_limit_date": "dt_limite_envio"
    ,"price": "preco"
    ,"freight_value": "vl_frete"
}, inplace=True)

df_orders_items.head()
```

### Base olist_sellers_dataset
*** 

* Informações dos vendedores. 
* Um pedido pode ter mais de um vendedor.

|Coluna | Descrição |
| :--- | :------- |
|seller_id|id do vendedor |
|seller_zip_code_prefix|prefixo do código postal do vendedor|
|seller_city|cidade vendedora|
|seller_state|uf vendedor|

```python
df_sellers = pd.read_csv("./base/olist_sellers_dataset.csv")
df_sellers.head()
```

```python
df_sellers.rename(columns={
    "seller_id": "id_vendedor"
    ,"seller_zip_code_prefix": "prefixo_cep_vendedor" 
    ,"seller_city": "cidade_vendedor"
    ,"seller_state": "uf _vendedor"
}, inplace=True)

df_sellers.head()
```

***

```python
df_reviews.rename(columns={
    "review_id": "id_revisao"
    ,"order_id": "id_pedido"
    ,"review_score": "pontuacao_revisão"
    ,"review_comment_title": "titulo_comentário"
    ,"review_comment_message": "mensagem_comentário"
    ,"review_creation_date": "dt_criacao_revisao"
    ,"review_answer_timestamp": "dt_carimbo_resposta"
}, inplace=True)

df_reviews.head()
```
